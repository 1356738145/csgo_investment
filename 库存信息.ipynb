```python
"""
Created by Charles-Deng
870055485@qq.com
Date: 2022/9/25 13:40
"""
import requests
import re
import time
import numpy as np
import pandas as pd
import csv
import json
import random
import math
#import talib

def get_price(id):
    url = "https://buff.163.com/api/market/goods/sell_order?game=csgo&goods_id="+ str(id) +"&page_num=1&sort_by=default"
    req = requests.get(url=url)
    req_data = json.loads(req.text)
    price_dic = req_data['data']['items'][0]
    return price_dic

df = pd.read_excel('C:/Users/Chandler/Desktop/csgo_ledger.xlsx')
sell_price = df["sell_price"]
sell_price_list = []
for price in sell_price:
    sell_price_list.append(price)
#print(sell_price_list)

id_ser = df['buff_id']
newest_price_list = []
time_list = []
count =0#sell_price索引对齐

for id in id_ser:
    if math.isnan(sell_price_list[count]):#不存在sell_price
        price_dic = get_price(id)
        time.sleep(random.uniform(0.5, 0.6))
        #time_stamp = price_dic['updated_at']
        price = round(float(price_dic['price']),1)
        #times = time.strftime("%Y/%m/%d-%H:%M:%S",time.localtime(time_stamp))
        #time_list.append(times)
        newest_price_list.append(price)
        print('id:{} is completed.'.format(id))
    else:#存在sell_price
        newest_price_list.append(sell_price_list[count])
    count = count + 1

new_df = df
new_df['newest_price'] = newest_price_list
new_df['sell_price'] = sell_price_list
#new_df['update_time'] = time_list
print(new_df)
buy_all = np.sum(new_df['buy_price'])
cash_out = np.sum(new_df['sell_price'])
now_all = np.sum(new_df['newest_price'])
print("Buy total cost: {},  cash out: {}, now total value: {}, basic earning: {}".format(str(buy_all), str(cash_out) ,str(now_all-cash_out), str(now_all-buy_all)))
#new_df.to_excel('csgo_ledger.xlsx',index=False)
```
